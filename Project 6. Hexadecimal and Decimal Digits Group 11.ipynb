```python
def hex2int(hex_str):
    """
    Converts a hexadecimal string (e.g., '1A', 'f3', '0') to its decimal integer value.
    Handles both uppercase and lowercase letters.
    Does NOT use built-in int(x, 16).
    """
    hex_str = hex_str.strip().upper()
    valid_hex_digits = "0123456789ABCDEF"

    for ch in hex_str:
        if ch not in valid_hex_digits:
            raise ValueError(f"Invalid hexadecimal digit: {ch}")

    decimal_value = 0
    power = 0

    for ch in reversed(hex_str):
        if '0' <= ch <= '9':
            digit_value = ord(ch) - ord('0')
        else:
            digit_value = ord(ch) - ord('A') + 10
        decimal_value += digit_value * (16 ** power)
        power += 1

    return decimal_value
```

```python
def int2hex(number):
    """
    Converts a non-negative integer to its hexadecimal string representation.
    Handles integers larger than 15 without using built-in hex().
    """
    if not isinstance(number, int):
        raise TypeError("Input must be an integer.")
    if number < 0:
        raise ValueError("Input must be a non-negative integer.")

    hex_digits = "0123456789ABCDEF"

    if number == 0:
        return "0"

    hex_str = ""
    while number > 0:
        remainder = number % 16
        hex_str = hex_digits[remainder] + hex_str
        number //= 16

    return hex_str
```

```python

def main():
    """
    Interactive program allowing user to convert between hexadecimal and decimal.
    Automatically retries on invalid input with an exit option.
    """
    while True:
        print("=== Hexadecimal <-> Decimal Converter ===")
        print("Choose conversion type:")
        print("Enter 1 to convert Hexadecimal to Integer")
        print("Enter 2 to covert Integer to Hexadecimal")

        choice = input("Enter 1 or 2: ").strip()

        if choice == "1":
            attempt_count = 0
            while True:
                hex_value = input("Enter a hexadecimal value (or type 'exit' to go back): ").strip()
                if hex_value.lower() == 'exit':
                    print("Returning to main menu...\n")
                    break
                try:
                    result = hex2int(hex_value)
                    print(f"The decimal equivalent of {hex_value} is {result}.")
                    break
                except ValueError as e:
                    attempt_count += 1
                    print(f"Error: {e}. Please try again.")
                    if attempt_count >= 3:
                        choice_exit = input("Too many invalid attempts.Do you want to exit? (yes/no): ").strip().lower()
                        if choice_exit == 'yes':
                            print("Exiting program...")
                            return
                        else:
                            attempt_count = 0
                    continue

        elif choice == "2":
            while True:
                int_value = input("Enter a non-negative integer (or type 'exit' to go back): ").strip()
                if int_value.lower() == 'exit':
                    print("Returning to main menu...\n")
                    break
                if not int_value.isdigit():
                    print("Error: Input must be a non-negative integer. Please try again.")
                    continue
                int_value = int(int_value)
                result = int2hex(int_value)
                print(f"The hexadecimal equivalent of {int_value} is {result}.")
                break

        else:
            print("Invalid choice. Please enter 1 or 2.\n")
            continue

        another = input("Do you want to perform another conversion? (yes/no): ").strip().lower()
        if another != 'yes':
            print("GOODBYE!")
            break


if __name__ == "__main__":
    main()
```
